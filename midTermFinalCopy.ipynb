```python
#import the required libraries
import urllib.request, urllib.parse, urllib.error
import json
import ssl
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
import re

delta = 7

dictOfQuakes = {}
dictOfQuakesLocation = {}
dictOfQuakesLngLat = {}
magList = []
duplicate = []
newMagList = []
i = 0
resList = []

dateDelta = datetime.now() - timedelta(days=delta)
dateDelta = dateDelta.strftime('%Y-%m-%d')

usgsJsonUrl = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=' + dateDelta + "&endtime"

print(usgsJsonUrl)

# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

while True:
    #print('Retrieving', usgsJsonUrl)
    
    #urlopen retrieves and returns a bytes object from the URL
    uh = urllib.request.urlopen(usgsJsonUrl, context=ctx)
    data = uh.read().decode()
    print('Retrieved', len(data), 'characters')
    
    #check to see if there is any data that was retrieved
    try:
        #if so store that data in this variable
        js = json.loads(data)
    except:
        js = None

    #print(json.dumps(js, indent=4))
    
    #json.loads() returns a list of dictionaries so we're using keys for dictionaries
    #and indexes for lists to retrieve the information we need.
    #here we're selecting the latitude and longitude to be printed
    
    listOfProps = js['features']
    
    for item in listOfProps:
        if item['properties']:
            if item['properties']['mag']:
                magList.append((item['properties']['mag'], item['properties']['url']))
                dictOfQuakesLocation[item['properties']['url']] = item['properties']['place']
                dictOfQuakesLngLat[item['properties']['url']] = item['geometry']['coordinates']
                
    magList.sort(reverse = True)
    
    while i <= 5:
        for x, y in magList:
            resList.append(y)
            newMagList.append(x)
        i += 1
    
    quakeOne =  "\"" + str(dictOfQuakesLocation[resList[1]]) + "\""
    quakeTwo = "\"" + str(dictOfQuakesLocation[resList[2]]) + "\""
    quakeThree = "\"" + str(dictOfQuakesLocation[resList[3]]) + "\""
    quakeFour = "\"" + str(dictOfQuakesLocation[resList[4]]) + "\""
    quakeFive = "\"" + str(dictOfQuakesLocation[resList[5]]) + "\""

    print(magList[0:5])
    topOfTheList = dictOfQuakesLngLat[resList[0]]
    lng = str(topOfTheList[0])
    lat = str(topOfTheList[1])
    quakeOneTitle =  str(dictOfQuakesLocation[resList[0]]) + " with a magnitude of " + str(newMagList[0])
    quakeOneLngLat = "lat " + lng + ", " + "lng " + lat

    break
    
    ##################################################################################################### 
        
data = """
<!DOCTYPE html>
<html>
  <head>
        <style>
@import "compass/css3";
<link href="https://fonts.googleapis.com/css?family=Montserrat" rel="stylesheet">
<link href="https://fonts.googleapis.com/css?family=Raleway" rel="stylesheet">
<link href="https://fonts.googleapis.com/css?family=Work+Sans" rel="stylesheet">
body {
  color: #666666;
}

h4{
font-family: 'Raleway', sans-serif;
font-weight: 50;
animation: 3s ease-out 0s 1 opacity;
color: black;
margin-left: 15%;
}

h2{
font-family: 'Work Sans', sans-serif;
animation: 2s ease-out 0s 1 opacity;
font-weight: 50;
color: black;
}

h3{
font-family: 'Raleway', sans-serif;
animation: 3s ease-out 0s 1 opacity;
font-weight: 50;
color: black;
margin-left: 15%;
}


.chart {
  display: table;
  table-layout: fixed;
  width: 60%;
  max-width: 700px;
  height: 200px;
  margin: 0 auto;
  background-image: linear-gradient(bottom, rgba(0, 0, 0, 0.1) 2%, transparent 2%);
  background-size: 100% 50px;
  background-position: left top;
  font-family: 'Raleway', sans-serif;
    color: ##666666;
    text-align: center;
    font-weight: 100;
}
.chart li {
  position: relative;
  display: table-cell;
  vertical-align: bottom;
  height: 200px;
  text-align: center;
}
.chart span {
  margin: 0 1em;
  display: block;
  background: #E98074;
  animation: draw 2s ease-in-out;
  text-align: center;
}

.chart span:hover {
background: #E85A4F;
transition: 0.9s;
}

.chart span:before {
  position: absolute;
  left: 0;
  right: 0;
  top: 100%;
  padding: 5px 1em 0;
  display: block;
  text-align: center;
  content: attr(title);
  word-wrap: break-word;
}

@keyframes draw {
  0% {
    height: 0;
  }
  50% {
    height: 10;
  }
  100% {
    height: 10;
  }
}

#map {
height: 200px;  
width: 70%; 
margin-left: auto;
margin-right: auto;
animation: 2s ease-out 0s 1 opacity;
top: 0; 
}

@keyframes opacity {
  0% {
        display: none;
        opacity: 0;
    }

    1% {
        display: block;
        opacity: 0;
    }

    100% {
        display: block;
        opacity: 1;
    }
}

@keyframes opacityOne {
  0% {
        display: none;
        opacity: 0;
    }

    80% {
        display: none;
        opacity: 0;
    }

    100% {
        display: block;
        opacity: 1;
    }
}

@keyframes slideUp {
  0% {
    transform: translateY(15%);
  }
  100% {
    transform: translateY(0);
  }
}

body {  

  animation: 1s ease-out 0s 1 slideUp;
  
  background-color: #EAE7DC;
  padding: 30px;
}

.center{
display: inline-block
}

p{
animation: 2s ease-out 0s 1 opacityOne;
font-size: 2em;
color: white;
}
    </style>
  </head>
  <body>
  <h2>This web app collects data from the usgs on earthquakes in the past seven days</h2>
    <h3>The map below displays the location of the earthquake with highest magnitude in the last seven days</h3>
    <h3>""" + quakeOneTitle + """</h3>
    <h4>""" + quakeOneLngLat + """</h4>
    <div id="map"></div>
    <script>
window.onload = display();

function display(){
    document.getElementById("map").style.opacity = "1";
}
function initMap() {

    var location = {lat: """ + lat + """, lng: """ + lng + """};
    var map = new google.maps.Map(document.getElementById('map'), {zoom: 4, center: location});
    var marker = new google.maps.Marker({position: location, map: map});
    
    }
    
    </script>

    <script async defer
    src="https://maps.googleapis.com/maps/api/js?key=AIzaSyBc_IgeMGIsaPTKIgWAk4qucFeX74rkXmU&callback=initMap">
    </script>
    <h3 style="padding-top:40px;">This chart contains the magnitude and location of the earthquakes with the next four highest magnitudes</h2>
<ul class="chart">
  <li>
    <span style="height:""" + str(newMagList[1] * 10) + """%" title=""" + quakeOne + """><div class="center"><p>""" + str(newMagList[1]) + """</p></div></span>
  </li>
  <li>
    <span style="height:""" + str(newMagList[2] * 10) + """%"title=""" + quakeTwo + """><div class="center"><p>""" + str(newMagList[2]) + """</p></div></span>
  </li>
  <li>
    <span style="height:""" + str(newMagList[3] * 10) + """%"title=""" + quakeThree + """><div class="center"><p>""" + str(newMagList[3]) + """</p></div></span>
  </li>
  <li>
    <span style="height:""" + str(newMagList[4] * 10) + """%"title=""" + quakeFour + """><div class="center"><p>""" + str(newMagList[4]) + """</p></div></span>
  </li>
</ul>    

  </body>
</html>
"""
###########################################################################################################

soup = BeautifulSoup(data, "html.parser")
#frame = soup.iframe

#print(data)

f = open('earthquakeWebApp.html','w')

#f.write(soup.prettify())
f.write(data)
f.close()

import webbrowser

# open an HTML file on my own (Windows) computer
url = "earthquakeWebApp.html"
webbrowser.open(url,new=2)
```
